```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/ageron/handson-ml2/master/datasets/housing/housing.csv"
data = pd.read_csv(url)
```

```python
data.head()
```

```python
print("Shape:", data.shape)
```

```python
print("Column names:", data.columns)
```

```python
print("Summary statistics:")
data.describe()
```

```python
print("Missing values:")
print(data.isnull().sum())
```

```python
data = data.dropna()
```

```python
print(data.isnull().sum())
```

```python
sns.histplot(data['median_house_value'], kde=True)
plt.xlabel("Median House Value")
plt.ylabel("Frequency")
plt.title("Distribution of Median House Value")
plt.show()
```

```python
correlation_matrix = data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="YlGnBu")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.boxplot(x=data['ocean_proximity'], y=data['median_house_value'])
plt.xlabel("Ocean Proximity")
plt.ylabel("Median House Value")
plt.title("Housing Prices by Ocean Proximity")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.scatter(data['longitude'], data['latitude'], c=data['median_house_value'], cmap='coolwarm', alpha=0.5)
plt.colorbar(label='Median House Value')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Geographical Distribution of Housing Prices")
plt.show()
```

```python
sns.pairplot(data[['median_house_value', 'total_rooms', 'housing_median_age', 'population']])
plt.show()
```

```python
sns.histplot(data['total_rooms'] / data['households'], bins=30, kde=True)
plt.xlabel("Average Rooms per Household")
plt.ylabel("Frequency")
plt.title("Distribution of Average Rooms per Household")
plt.show()
```
